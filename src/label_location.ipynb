```python
from PIL import Image
import matplotlib.pyplot as plt
import shutil
import os
import re
import csv
import cv2 as cv
from google.colab.patches import cv2_imshow
```

```python
txt_files = ["/content/drive/MyDrive/Xu ly anh /TRECVID/DVU2022/queries/Calloused_Hands.entity.types.txt",
          "/content/drive/MyDrive/Xu ly anh /TRECVID/DVU2022/queries/Liberty_Kid.entity.types.txt",
          "/content/drive/MyDrive/Xu ly anh /TRECVID/DVU2022/queries/chained_for_life.entity.types.txt",
          "/content/drive/MyDrive/Xu ly anh /TRECVID/DVU2022/queries/like_me.entity.types.txt",
          "/content/drive/MyDrive/Xu ly anh /TRECVID/DVU2022/queries/little_rock.entity.types.txt",
          "/content/drive/MyDrive/Xu ly anh /TRECVID/DVU2022/queries/losing_ground.entity.types.txt"]


image_folder =["/content/drive/MyDrive/Xu ly anh /TRECVID/DVU2022/queries/images/Calloused_Hands",
               "/content/drive/MyDrive/Xu ly anh /TRECVID/DVU2022/queries/images/Liberty_Kid",
               "/content/drive/MyDrive/Xu ly anh /TRECVID/DVU2022/queries/images/chained_for_life",
               "/content/drive/MyDrive/Xu ly anh /TRECVID/DVU2022/queries/images/like_me",
               "/content/drive/MyDrive/Xu ly anh /TRECVID/DVU2022/queries/images/little_rock",
               "/content/drive/MyDrive/Xu ly anh /TRECVID/DVU2022/queries/images/losing_ground"]

classes_folder = "/content/drive/MyDrive/Xu ly anh /Group_Thanh_Ha/few_shot dataset"
# Tạo dictionary từ file txt
data = {}
for i in txt_files:
    file_name = i.split("/")[-1]
    movie_name = file_name.split(".")[0]
    with open(i, "r") as f:
        for line in f:
            key, value = line.strip().split(' : ')
            data[key] = value
            if value == "Location":
                class_name = f"{movie_name}_{key}"
                class_name = re.sub("'", "_", class_name.lower())
                #Tạo thư mục mới
                classes_path = os.path.join(classes_folder, class_name)
                os.makedirs(classes_path, exist_ok=True)

                for h in image_folder:
                    if (os.path.basename(h) == movie_name):

                        # Duyệt qua tất cả các file trong thư mục
                        for k in os.listdir(h):
                            file_path = os.path.join(h, k)

                            # Kiểm tra nếu là file ảnh
                            if os.path.isfile(file_path) and (file_path.endswith(ext) for ext in ['.png', '.jpg', '.jpeg']):
                                img_name, form = os.path.splitext(os.path.split(file_path)[1])
                                img_name = img_name.rsplit("_", 1)[0]
                                img_name = re.sub("'", "_", img_name.lower())
                                key = re.sub("'", "_", key.lower())
                                #Nếu tên ảnh trùng tên key trong file txt thì copy ảnh đó vào class mới
                                if(img_name == key):
                                    new_file_path = os.path.join(classes_path, k)
                                    shutil.copy(file_path, new_file_path)
```
